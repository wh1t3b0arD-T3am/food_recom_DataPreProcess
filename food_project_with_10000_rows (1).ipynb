```python
lst = []               
for i in data.ingredients:
    lst.append(i)         #putting all ingredients values into the list 'lst'
    
    

#for nonveg ---> 1   for veg ---> 0

for i in range(len(lst)):
    count = 0
    j = lst[i].split(" ")   #getting words for each list values
    for k in j:
        if  k in non_veg:
            data['type'][i] = 1
            break
        else:
            count = count+1      #traversing through j into the last word, if last word is also not in 'non_veg list' then 0
            if(count == len(j)):
                data['type'][i] = 0
                break
            
```

```python
# Lemmatization
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
for i in range(len(data.ingredients)):
    words = nltk.word_tokenize(data.ingredients[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    data.ingredients[i] = ' '.join(words) 
```

```python
%%time
for i in range(len(data.instructions)):
    words = nltk.word_tokenize(data.instructions[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    data.instructions[i] = ' '.join(words)
```

```python
data['unique_ingredients'] = ""    #creating new column
```

```python
#creating set for each 'ingredients'

for i in range(len(data.ingredients)):
    unique = set()
    for j in data.ingredients[i].split(" "):
        unique.add(j)
    data['unique_ingredients'][i] = unique
    
```

```python
data['unique_ingredients'] = data['unique_ingredients'].apply(lambda x: list(x)) # set to list
```

```python
data['calories'] = np.nan    #creating new calories column
```

```python
data.head()
```

```python
#calculating calorie count from 'unique_ingredients' column

%%time
for i in range(len(data['unique_ingredients'])):
    lst = data['unique_ingredients'][i]
    cal = 0
    for j in lst:
        for k in range(len(calories.Food)):
            if(j == calories.Food[k]):
                cal = cal + calories.calories[k]
    data['calories'][i] = cal            
                
```

```python
data['calories'] = data.calories.astype(int)  #float to int
```
