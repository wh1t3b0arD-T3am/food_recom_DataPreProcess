```python
for i in type_change:
    df[i] = df[i].str.replace(r'[^\d.]+', '')   #removing all alphabets, junk char
```

```python
from sklearn.experimental import enable_iterative_imputer    # to handle missing nan datas
```

```python
from sklearn.impute import IterativeImputer
```

```python
impute = IterativeImputer()
impute.fit_transform(df)
```

```python
df[:]= impute.transform(df)
```

```python
df.isna().sum()   #no nan values left
```

```python
df.head()  #but after using iterativeImputer the nan values were changed to -ve values which is 
#incorrect so we will perform knnImputer
```

```python
from sklearn.impute import KNNImputer
```

```python
for i in type_change:
    knn_data[i] = knn_data[i].str.replace(r'[^\d.]+', '')
```

```python
impute_knn = KNNImputer(n_neighbors = 3)
impute_knn.fit_transform(knn_data)
```

```python
knn_data[:]= impute_knn.transform(knn_data)
```

```python
knn_data   #no more -ve values
```

```python
for i in type_change:
    knn_data[i] = knn_data[i].astype('int64')
```
