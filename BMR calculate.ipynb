```python
'''1)	BMR for Men = 66.47 + (13.75 * weight [kg]) + (5.003 * size [cm]) − (6.755 * age [years])
2)	BMR for Women = 655.1 + (9.563 * weight [kg]) + (1.85 * size [cm]) − (4.676 * age [years])


Example – 

Gender- Male,   Age – 23 , Height – 180cm , Weight-80kg
BMR for Men = 66.47 + (13.75 *80) + (5.003 * 180) − (6.755 * 23) =  1658.39 Kcal/Day 


But if……


Little/no exercise: BMR * 1.2 = 1658.39*1.2 = 1990 Kcal/Day
Light exercise: BMR * 1.375 = 1658.39*1.375 = 2280 Kcal/Day
Moderate exercise (3-5 days/wk): BMR * 1.55 = 1658.39*1.55 = 2570.50 Kcal/Day
Very active (6-7 days/wk): BMR * 1.725 = 1658.39*1.725 =  2860.72 Kcal/Day
Extra active (very active & physical job): BMR * 1.9 = 3150.94 Kcal/Day

'''
```

```python
df['height'] = df['height'].apply(lambda x: x*2.54)    # making from inch to cm
```

```python
df['weight'] = df['weight'].apply(lambda x: x/2.205)    # making from pound to kg
```

```python
df.head()
```

```python
df['bmr'] = 66.47 + (13.75 * df['weight']) + (5.003 * df['height']) - (6.755 * df['age'])      #since no gender specfied so using the formula for male
```

```python
df['new_bmr_after_excercise'] = ""
```

```python
'''
Little/no exercise: BMR * 1.2 = 1658.39*1.2 = 1990 Kcal/Day
Light exercise: BMR * 1.375 = 1658.39*1.375 = 2280 Kcal/Day
Moderate exercise (3-5 days/wk): BMR * 1.55 = 1658.39*1.55 = 2570.50 Kcal/Day
Very active (6-7 days/wk): BMR * 1.725 = 1658.39*1.725 =  2860.72 Kcal/Day
Extra active (very active & physical job): BMR * 1.9 = 3150.94 Kcal/Day
'''
```

```python
#to calculate bmr calorie value wrt new_excercise

for i in df['new_excercise']:
    if i == 'sedentary little':
        df['new_bmr_after_excercise'] = df['bmr'] * 1.2
    elif i == 'lightly active':
        df['new_bmr_after_excercise'] = df['bmr'] * 1.375
    elif i == 'moderatetely active':
        df['new_bmr_after_excercise'] = df['bmr'] * 1.55
    elif i == 'very active':
        df['new_bmr_after_excercise'] = df['bmr'] * 1.725
    else:
        df['new_bmr_after_excercise'] = df['bmr'] * 1.9
                 
```
